```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
land = ""
rb = ""
data = []
ofile = open("2017.csv","r")
header = ofile.readline()
for line in ofile:
    if line[0].isdigit():
        data.append([2017,land, rb] + line.replace("\n","").split(";"))
    elif line.startswith("RB"):
        rb = line.split(" ")[1].replace("\n","").split(";")[0]
    else:
        land = line.replace("\n","").split(";")[0]
ofile.close()
        
df = pd.DataFrame(data, columns = ["year","land", "rb"] + header.replace("\n","").split(";"))
```
